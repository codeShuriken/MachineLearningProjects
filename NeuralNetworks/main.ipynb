# PART0: IMPORTS

```python
import numpy as np
import pandas as pd
import util_mnist_reader
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten
```

# PART1: SINGLE LAYER NEURAL NETWORK FROM SCRATCH

```python
#Load Dataset
X_train, y_train = util_mnist_reader.load_mnist('data/fashion', kind='train')
X_test, y_test = util_mnist_reader.load_mnist('data/fashion', kind='t10k')
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=42)

#Normalize the dataset
X_train = X_train/255
X_val = X_val/255
X_test = X_test/255

#Convert to categorical 
y_train = to_categorical(y_train, num_classes=10)
y_val = to_categorical(y_val, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

labelNames = ["top", "trouser", "pullover", "dress", "coat", "sandal", "shirt", "sneaker", "bag", "ankle boot"]

#print(X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape)

def init_weights(d1, d2):
    """This function initializes the starting weights as zeroes.

    Args:
        fea: The number of features
    Returns:
        The zero initialized weights: w and b

    """
    w = np.random.randn(d1, d2)
    b = np.random.randn(d1, 1)
    return w, b

def sigmoid(z):
    """This function performs sigmoid function for the  output of the 
    basis function on every single example.

    Args:
        z: The results of the basis function
    Returns:
        The output after applying basis function

    """
    a = 1 / (1 + np.exp(-z))
    return a


def softmax(z):
    """This function performs softmax function for the output of the 
    basis function on every single example.
    
    Args: 
        z: The results of the basis function
    Returns:
        The output after applying softmax function
    """
    a = (np.exp(z)) / (np.exp(z).sum(axis=0))
    return a

def plot_graph(X1, Y1, Y2, Xlabel, Ylabel):
    """This function plots the graph for the given X1 and Y1.

    Args:
        X1: The X-axis values
        Y1: The Y-axis values
        title: The tile of the graph
        Xlabel: The label of X-axis
        Ylabel: The label of Y-axis
    """
    plt.figure(figsize=(10,3))
    
    plt.subplot(1, 2, 1)
    plt.plot(X1, Y1, color='blue')
    plt.xlabel(Xlabel)
    plt.ylabel('Training ' + Ylabel)
    
    plt.subplot(1, 2, 2)
    plt.plot(X1, Y2, color='green')
    plt.xlabel(Xlabel)
    plt.ylabel('Cross-Validation ' + Ylabel)
    plt.show()


def predict1(a2, y):
    y_cap = []
    tmp = []
    for row in a2:
        maximum = max(row)
        for ind, ele in enumerate(row):
            if (ele == maximum):
                tmp.append(1)
            else:
                tmp.append(0)
        y_cap.append(tmp)
        tmp = []
    y_cap = np.array(y_cap)
    return classification_report(y_cap.argmax(axis=1), y.argmax(axis=1), target_names=labelNames)

def predict(a2, y):
    """This function predicts the models accuracy.

    Args:
        a2: The predicted values
        y: The actual values
    Returns:
        accuracy
    """
    y_cap = []
    tmp = []
    for row in a2:
        maximum = max(row)
        for ind, ele in enumerate(row):
            if (ele == maximum):
                tmp.append(1)
            else:
                tmp.append(0)
        y_cap.append(tmp)
        tmp = []
    Ntrue = 0
    for a1, a2 in zip(y, y_cap):
        if (np.array_equal(a1, a2)):
            Ntrue += 1
    accuracy = Ntrue / y.shape[0]
    return accuracy
    
    
    
def model(X_train, y_train, W1, b1, W2, b2, learning_rate, epochs, X_val, y_val):
    """This function performs neural network with one hidden layer on the input data using 
    Cross Entropy loss function and gradient descent optimizer.

    Args:
        X_train: The training dataset
        y_train: The output for the training dataset
        W1: The initial weights W1
        b1: The initial bias b1
        W2: The output layer weights
        b2: The output layer bias
        learning_rate: The learning rate for gradient descent
        epochs: The number of iterations to update the weights
        X_val: The cross-validation dataset
        y_val: The outputs for cross-validation dataset
    Returns:
        W1, W2, b1 and b2: The final weights and bias.
        Lists containing accuracies for each iteration for both training and validation datasets. 

    """
    
    losses, losses_val, accuracies, accuracies_val = [], [], [], []
    losses1, losses_val1, accuracies1, accuracies_val1 = [], [], [], []
    #Split the dataset
    X1 = np.array_split(X_train, 30)
    y1 = np.array_split(y_train, 30)
    #Iterate over the number of epochs
    for i in range(epochs):
        #Train the data in batches
        for X, y in zip(X1, y1): 
            m = X.shape[0]
            z1 = np.dot(W1, X.T) + b1
            a1 = sigmoid(z1)

            z2 = np.dot(W2, a1) + b2
            a2 = softmax(z2)
            a2 = a2.T
            #print(a2.shape)

            #Cross-Entropy loss function
            L = -1/m * np.sum(np.log(a2) * y) 
            losses.append(L)

            #Calculate loss on Validation set
            z1_val = np.dot(W1, X_val.T) + b1
            a1_val = sigmoid(z1_val)
            z2_val = np.dot(W2, a1_val) + b2
            a2_val = softmax(z2_val)
            a2_val =a2_val.T
            L_val = -1/X_val.shape[0] * np.sum(np.log(a2_val) * y_val) 
            losses_val.append(L_val)


            #Perform gradient descent to update weights and bias
            dz2  = a2 - y
            dw2 = 1/m * np.dot(dz2.T, a1.T)
            db2 = 1/m * np.sum(dz2.T, axis=1, keepdims=True)

            tmp = sigmoid(z1) * (1 - sigmoid(z1))
            dztemp =  np.dot(dz2, W2)
            dz1 =  tmp * dztemp.T
            dw1 = 1/m * np.dot(dz1, X)
            db1 = 1/m * np.sum(dz1, axis=1, keepdims=True)
            
            #Update weights and bias
            W1 = W1 - learning_rate * dw1
            b1 = b1 - learning_rate * db1
            W2 = W2 - learning_rate * dw2
            b2 = b2 - learning_rate * db2
            
            accuracy = predict(a2, y)
            accuracies.append(accuracy) 
      
            accuracy_val = predict(a2_val,y_val)
            accuracies_val.append(accuracy_val)
        
        #Take the average loss and accuracy of each batch
        losses1.append(sum(losses) / len(losses))        
        losses_val1.append(sum(losses_val) / len(losses_val))
        
        accuracies1.append(sum(accuracies) / len(accuracies))
        accuracies_val1.append(sum(accuracies_val) / len(accuracies_val))
        
        losses, losses_val, accuracies, accuracies_val = [], [], [], []
        
        #Display results
        if (i+1)%1 == 0:
            print('After ', i+1, ' epochs, loss: ', losses1[i] ,'accuracy: ', accuracies1[i],
              'cross_validation loss: ', losses_val1[i] , 'cross_validation accuracy: ', accuracy_val)
          
    return W1, b1, W2, b2, losses1, losses_val1, accuracies1, accuracies_val1

#Initialize Weights
hidden_units = 128
output_units = 10
W1, b1 = init_weights(hidden_units, X_train.shape[1])
W2, b2 = init_weights(output_units, hidden_units)
print(W1.shape, b1.shape, W2.shape, b2.shape)

learning_rate = 0.3
epochs = 5000

W1, b1, W2, b2, losses, losses_val, accuracies, accuracies_val = model(X_train, y_train, W1, b1, W2, b2, learning_rate, epochs, X_val, y_val)

#Plot the results
epochs1 = [i for i in range(1, epochs+1)]
plot_graph(epochs1, losses, losses_val, 'Epochs', 'Loss')
plot_graph(epochs1, accuracies, accuracies_val, 'Epochs', 'Accuracy') 

#Test our model
z1_test = np.dot(W1, X_test.T) + b1
a1_test = sigmoid(z1_test)
z2_test = np.dot(W2, a1_test) + b2
a2_test = softmax(z2_test)
a2_test = a2_test.T
test_accuracy = predict1(a2_test, y_test)
print('Test_Accuracy: \n', test_accuracy)
```

# PART2: MULTI-LAYER NEURAL NETWORK USING KERAS

```python
def plotGraph(history):
    plt.figure(figsize=(10,3))
    
    # Plot training & validation accuracy values
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Cross-Validation'], loc='upper left')
    
    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Cross-Validation'], loc='upper left')

    plt.show()
```

```python
model = Sequential()
model.add(Dense(units=512, input_dim = 784, activation='relu'))
model.add(Dense(units=512, activation='relu'))
model.add(Dense(units=512, activation='relu'))
model.add(Dense(units=10, activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs = 100)
```

```python
plotGraph(history)
preds = model.predict(X_test)
print(classification_report(y_test.argmax(axis=1), preds.argmax(axis=1), target_names=labelNames))
#model.summary()
```

# PART3: CONVOLUTION NEURAL NETWORK USING KERAS

```python
X_train1 = X_train.reshape((60000, 28, 28, 1))
X_val1 = X_val.reshape((5000, 28, 28, 1))
X_test1 = X_test.reshape((5000, 28, 28, 1))
#print(X_train1[0])

model1 = Sequential()
model1.add(Conv2D(28, (3, 3), activation='relu', input_shape=(28,28, 1)))
model1.add(MaxPool2D(2, 2))
model1.add(Conv2D(56, (3, 3), activation='relu'))
model1.add(MaxPool2D(2, 2))
model1.add(Flatten())
model1.add(Dense(units=512, activation='relu'))
model1.add(Dense(units=512, activation='relu'))
model1.add(Dense(units=10, activation='softmax'))
model1.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
history1 = model1.fit(X_train1, y_train, validation_data=(X_val1, y_val), epochs = 100)
```

```python
plotGraph(history1)
preds = model1.predict(X_test1)
print(classification_report(y_test.argmax(axis=1), preds.argmax(axis=1), target_names=labelNames))
#model1.summary()
```
